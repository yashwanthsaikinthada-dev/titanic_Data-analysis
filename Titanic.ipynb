```python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px
import plotly.graph_objs as go
from IPython.display import IFrame


# Create the app
app = Dash(__name__)

# ---------------- Graphs ---------------- #

# Pie chart
survived_counts = df['Survived'].value_counts()
pie_chart = go.Figure(
    data=[go.Pie(labels=["Not Survived", "Survived"], values=survived_counts, hole=0.3)],
)
pie_chart.update_layout(title="Survived vs Not Survived")

# Histogram + Line Plot
counts, bins = np.histogram(df["Age"].dropna(), bins=20)
bin_centers = (bins[:-1] + bins[1:]) / 2
histogram = go.Figure()
histogram.add_trace(go.Bar(x=bin_centers, y=counts, marker_color='orange', name='Age'))
histogram.add_trace(go.Scatter(x=bin_centers, y=counts, mode='lines+markers', name='Trend'))
histogram.update_layout(title="Age Distribution", xaxis_title="Age", yaxis_title="Frequency")

# Bar chart
val = df['Pclass'].value_counts().index.astype(str)
count = df['Pclass'].value_counts().values
bar_chart = go.Figure(data=[go.Bar(x=val, y=count, marker_color=['red', 'green', 'blue'])])
bar_chart.update_layout(title="Passenger Class Distribution", xaxis_title="Class", yaxis_title="Count")

# Pclass vs Survived
pclass_survived = df.groupby(['Pclass', 'Survived']).size().reset_index(name='Count')
countplot = px.bar(pclass_survived, x="Pclass", y="Count", color="Survived", barmode="group",
                   title="Pclass vs Survived")

# Heatmap
corr_matrix = df[['Survived', 'Pclass', 'Age', 'Fare']].corr()
heatmap = go.Figure(data=go.Heatmap(
    z=corr_matrix.values,
    x=corr_matrix.columns,
    y=corr_matrix.columns,
    colorscale='RdBu',
    zmin=-1,
    zmax=1,
    colorbar=dict(title="Correlation")
))
heatmap.update_layout(title="Correlation Heatmap")

# Scatter Plot
scatter = px.scatter(df, x='Age', y='Fare', color='Pclass', title="Age vs Fare")

# Scatter Matrix
pairplot = px.scatter_matrix(df[['Fare', 'Age']])
pairplot.update_layout(title="Scatter Matrix: Fare vs Age")

# ---------------- Layout ---------------- #

app.layout = html.Div([
    html.H1("Titanic Data Dashboard", style={'textAlign': 'center', 'color': '#003366'}),
    dcc.Tab(label='Dashboard', children=[
    html.Div([

        html.Div([
            html.Div(dcc.Graph(figure=pie_chart), style={'width': '48%', 'padding': '1%'}),
            html.Div(dcc.Graph(figure=histogram), style={'width': '48%', 'padding': '1%'}),
        ], style={'display': 'flex', 'justifyContent': 'space-between'}),

        html.Div([
            html.Div(dcc.Graph(figure=heatmap), style={'width': '32%', 'padding': '1%'}),
            html.Div(dcc.Graph(figure=bar_chart), style={'width': '32%', 'padding': '1%'}),
            html.Div(dcc.Graph(figure=countplot), style={'width': '32%', 'padding': '1%'}),
        ], style={'display': 'flex', 'justifyContent': 'space-between'}),

        html.Div([
            html.Div(dcc.Graph(figure=scatter), style={'width': '48%', 'padding': '1%'}),
            html.Div(dcc.Graph(figure=pairplot), style={'width': '48%', 'padding': '1%'}),
        ], style={'display': 'flex', 'justifyContent': 'space-between'}),

    ])
]),
], style={'padding': '10px', 'backgroundColor': '#f9f9f9'})


# Then visit http://127.0.0.1:8050/

app.run(debug=True, port=8050)
```
